```python
import seaborn as sn
plt.style.use('seaborn')
sales_10.plot(kind ='bar', figsize =(10, 5), fontsize = 10, color = 'black')
plt.xlabel("States", fontsize = 10)
plt.ylabel("Total Sales", fontsize = 10)
plt.title("Top 10 States by Sales", fontsize = 10)
plt.show()
```

```python
Central = list(d.groupby("Region"))[0][1]
```

```python
Category_Furniture = list(list(d.groupby("Region"))[0][1].groupby("Category"))[0][1]
```

```python
plt.figure(figsize = (10, 5))
sns.set(font_scale = 1)
sns.heatmap(Category_Furniture.corr(), annot = True, cmap = "Reds")
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.set(font_scale = 1, palette = "viridis")
sns.barplot(data = Category_Furniture , x = "State", y = "Profit", hue = "Sub-Category")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Sub Category)", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.set(font_scale = 1, palette = "viridis")
sns.barplot(data = Category_Furniture, x = "State", y = "Profit", hue = "Segment")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Segment)", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.set(font_scale = 1, palette = "viridis")
sns.barplot(data = Category_Furniture , x = "State", y = "Discount", hue = "Sub-Category")
plt.title("Discounts provided by each state", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.set(font_scale = 1)
sns.lmplot(data = Category_Furniture, x = "Discount", y = "Sales", aspect = 1, height = 5, col ="Sub-Category", col_wrap= 2)
plt.show()
```

```python
sns.set(font_scale = 1)
sns.jointplot(data = Category_Furniture, x = "Discount", y ="Sales", height = 5, kind = "reg")
plt.show()
```
